```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np

# 載入資料
data = load_breast_cancer()
X = data.data
y = data.target

# 查看資料說明與 shape
# print(data.DESCR)
print(f"X shape = {X.shape}, y shape = {y.shape}")

# 可以用 pandas DataFrame 方便觀察
df = pd.DataFrame(X, columns=data.feature_names)
df["target"] = y
# print(df.head())

m, n = X.shape
lambda_ = 0
learning_rate = 0.1
iteration = 1000
w = np.random.randn(30,)
b = np.random.randn()
costs = []

def sigmoid(w, X, b):
    z = np.dot(X, w) + b
    return 1 / (1 + np.exp(-z))
    
def compute_cost(w, b, X, y, lambda_):
    m, n = X.shape
    return (-1/m)*np.sum(y * np.log(sigmoid(w, X, b)) + (1-y) * np.log(1-sigmoid(w, X, b))) \
    + (lambda_/(2*m) * np.sum(w**2))
    
for i in range(iteration):
    y_pred = sigmoid(w, X, b)
    cost = compute_cost(w, b, X, y, lambda_)
    costs.append(cost)
    w -= learning_rate * ((1/m) * np.dot(X.T, (y_pred - y)) + (lambda_/m) * w)
    b -= learning_rate * ((1/m) * np.sum(y_pred - y))
    if i % (iteration//10) == 0:
        print(f"iteration = {i}, cost = {cost}, w = {w}, b = {b}")

print(f"Final model : w = {w}, b = {b}")


```
